```python
# importing necessarily libraries
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
# OLX main searching page url 
start_url = 'https://www.olx.com.eg/jobs/guards-security/'
```

```python
def pages_counter(url):
    # This function creates a list of all search pages.
    
    # Opening page..
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2228.0 Safari/537.36',}
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content)
    # Fetching last page count..
    pages_count = int(soup.find_all(class_='item fleft')[-1].find('span').string)
    # Creating a list of all pages urls..
    lst=[]
    for i in range(1,pages_count+1):
        lst.append(url+'?page='+str(i))
    return lst
```

```python
def page_ads_urls(url):
    # This function loops through page to get ads urls

    # Opening page..
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2228.0 Safari/537.36',}
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content)
    # Fetching each ad url..
    links = soup.find(class_='rel listHandler').find_all(class_='ads__item__ad--title')
    lst = []
    for link in links:
        lst.append(link.get('href'))
    return lst
```

```python
def extractor(url):
    # This function is responsible of fetching ad data and return it as a dictionary.
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/41.0.2228.0 Safari/537.36',}
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content,'lxml')

    title = soup.find(class_="brkword lheight28").string.replace("\n","").replace("\t","")
    location = soup.find(class_='c2b small').string.replace("\n","").replace("\t","")
    id = soup.find(class_='rel inlblk').string.replace("\n","").replace("\t","").replace(" ","")
    timestamp = soup.find(class_="pdingleft10 brlefte5").contents[0].replace("\n","").replace("\t","")
    user_name = soup.find(class_="user-box__info__name").string
    user_url = soup.find(class_='user-box__links__link').get('href')

    description=""
    for i in soup.find(class_ = "clr descriptioncontent marginbott20").find(id="textContent").p.stripped_strings:
        description+=i+" "
    try:
        user_type = soup.find(class_ = "clr descriptioncontent marginbott20").find(string='نوع المُعلن').parent.parent.a.string.replace("\n","").replace("\t","")
    except:
        user_type=""
        
    try:
        exp_level = soup.find(class_ = "clr descriptioncontent marginbott20").find(string='المستوى التعليمي').parent.parent.a.string.replace("\n","").replace("\t","")
    except:
        exp_level=""
        
    try:
        edu_level = soup.find(class_ = "clr descriptioncontent marginbott20").find(string='مستوى الخبرة').parent.parent.a.string.replace("\n","").replace("\t","")
    except:
        edu_level=""
        
    try:
        emp_type = soup.find(class_ = "clr descriptioncontent marginbott20").find(string='نوع العمل').parent.parent.a.string.replace("\n","").replace("\t","")
    except:
        emp_type=""
        
    try:
        salary = soup.find(class_ = "clr descriptioncontent marginbott20").find(string='الراتب').parent.parent.strong.string.replace("\n","").replace("\t","")
    except:
        salary=""
    
    values = {'title': title,
              'location': location,
              'id': id,
              'timestamp': timestamp, 
              'user_name': user_name,
              'user_url': user_url,
              'description': description,
              'user_type': user_type,
              'exp_level': exp_level,
              'edu_level': edu_level,
              'emp_level': emp_type,
              'salary': salary
             }
    print('Success .. ')
    return values
```

```python
# Creating a list of pages urls..
pages_urls = pages_counter(start_url)
all_links=[]
print('Starting ..')
for i,link in enumerate(pages_urls):
    try:
        links = page_ads_urls(link)
        all_links.extend(links)
        print("Page ",i+1," is completed")
    except:
        print("Failure in page ",i+1)
```

```python
# Looping through urls to scrape the data..
df = []
for link in all_links:
    dict = extractor(link)
    df.append(dict)

# Convert the list of data into a data frame..
df = pd.DataFrame(df)
df
```

```python
# Export data into a CSV file
df.to_csv('olx_data.csv')
```
